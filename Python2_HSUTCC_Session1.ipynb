Factorial
10! = 10 * 9 * ... * 1
10! = 10 * 9!

9! = 9 * 8!

2! = 2 * 1!
1! = 1

```factorial(10) = 10 * factorial(9)```


````python
# 5! = 5 x 4 x 3 x 2 x 1
def recursive_factorial(x):
    if x == 1:
        return 1
    return x * recursive_factorial(x-1)

recursive_factorial(10)
````

````python
# 5! = 5 x 4 x 3 x 2 x 1
def interative_factorial(x):
    result = 1

    for number in range(1, x + 1):
        result *= number
    
    return result

interative_factorial(10)
````

````python
%%timeit
import time

def recursive_factorial(x):
    if x == 1:
        return 1
    return x * recursive_factorial(x-1)

recursive_factorial(500)

t1 = time.time() 
#measures time in seconds, but with limited precision (only about ~1 millisecond accuracy).
result = recursive_factorial(6)
t2 = time.time()
print('Time usage for recursive_factorial 6: ', t2-t1)
result = recursive_factorial(501)
t3 = time.time()
print('Time usage for recursive_factorial 500: ', t3-t2)


# time.time() is simple, but low precision.
# %%timeit is accurate, repeats runs many times, and gives results in microseconds (s).
# Recursive factorials for small numbers run too fast for manual timing.
````

````python
# Mutable vs Immutable
x = 2
print(id(x))

x += 1 # x = 3
print(id(x))

'''
int is immutable  it cannot be changed in place.
When you do x = 3, Python creates a new object (3) and makes x point to it.
The old value (2) still exists somewhere else, but x no longer refers to it.
Reassigning x changes what it points to, not the data itself.
'''
````

````python
a = [1, 2, 3]
print(id(a))

a.append(4)
print(id(a))

# same id
'''list is mutable  you can modify its contents.
When you call append(4), the same list object in memory is changed.
You did not reassign a; you modified it.'''
````

In function, you can **edit** but **not modify** global variables!

In a function, you can edit (mutate) a global variable,
but you cannot reassign it to something new unless you use global.

````python
def add_to_list(num):
	very_important_list.append(num)

add_to_list(4)
print(very_important_list)

'''
The function directly edits the global list very_important_list.
append() changes the list in place (mutates it).
This means your function depends on and alters a global variable.

Problems
- the function changes data outside its scope.
- Harder to debug if something goes wrong later.
- Not reusable = it only works with that specific global variable.
'''
````

Let's fix that.

````python
def add_to_list_v2(num, target):
    target = target.copy()
    target.append(num)
    return target

print(add_to_list_v2(5, very_important_list))
print(very_important_list)

'''
The function receives the list as an argument (good practice).
Then it creates a copy of that list.
append() modifies only the copied list, not the original one.
The function returns the new list, leaving the original unchanged.
'''
````

### Sidenote: List as default argument

````python
def add_to_list_2_5(target=[i + 1 for i in range(3)]):
    print(target)

add_to_list_2_5()
````

````python
def add_to_list_3(num, target=[]):
    target.append(num)
    print(target)

'''
target=[]  a default empty list is created once when the function is defined, not every time you call the function.
target.append(num)  modifies the list in place.
print(target)  shows the current state of the list.
'''

add_to_list_3(1)
add_to_list_3(1) # target = [1]
add_to_list_3(1)
add_to_list_3(1)
add_to_list_3(1)
add_to_list_3(1, [2])
add_to_list_3(1)
````

````python
def add_to_list_3(num, target=[]):
    target = target.copy() 
    # creates a new list every time the function is called.
    # So even if you used the default list, you dont modify the original default list.
    target.append(num)
    # Appending happens only on the copied list, which is local to this call.
    print(target)
    # Each call prints [1] because:
    # The default list is never modified.
    # Every call creates a fresh copy to work with.

add_to_list_3(1)
add_to_list_3(1) # target = [1]
add_to_list_3(1)
add_to_list_3(1)
add_to_list_3(1)
add_to_list_3(1, [2])
add_to_list_3(1)
````

The correct way

````python
def add_to_list_3_corrected(num, target = None): #Parameters: (num, target=None)
    # if target is None:
    #     target = []
    target = target or []
    target.append(num)
    print(target)

'''
target=None  default argument is immutable (None)
target = target or []  if target is None (or any falsy value), create a new list.
target.append(num)  append to the newly created list
'''

add_to_list_3_corrected(1) # target = None  []  append 1  [1]
add_to_list_3_corrected(1) # new [] created  append 1  [1]
add_to_list_3_corrected(1) # new [] created  append 1  [1]
add_to_list_3_corrected(1)
add_to_list_3_corrected(1)
add_to_list_3_corrected(1, [2]) # target is [2]
add_to_list_3_corrected(1)
````

1. Write a Python function to find the max of any numbers of arguments (don't use the build in function `max()`)

````python
# You work here
def find_max_number(*args):
    if not args:
        return None

    maximum = args[0]
    for num in args[1:]:
        if num > maximum:
            maximum = num
    return maximum

print(find_max_number())
print(find_max_number(-5, -8, 0))
````

4. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

````python
# Your work here
def count_upper_lower(the_string):
    upper_count = 0
    lower_count = 0

    for char in the_string:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1
    return upper_count, lower_count

print(count_upper_lower('Hello'))
````
