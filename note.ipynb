```python
contact = ['Ve', 'Due']

def sms_decorator(func):
    def inner(contact):
        for each_contact in contact:
            print(f'send sms to {each_contact}')  # extra behavior before
        func(contact)  # call the original function
    return inner

@sms_decorator
def send_notice(contact):
    print("Notification sent!")
```

```python
import time

def timethis(func):
    @functools.wraps(func)
    def inner(*args, **kwargs):
        print(func.__name__, end='...')

        start_time = time.time()
        result = func(*args, **kwargs)
        elapse_time = time.time() - start_time

        print(elapse_time)
        return result
    
    return inner
```

```python
def once(func):
    @functools.wraps(func)
    def inner(*args, **kwargs):

        if not inner.called:
            func()
            inner.called = True

    inner.called = False

    return inner
```

```python
@once
def print_3():
    print(3)

print(3)
print(3)
```
