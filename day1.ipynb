# list comprehension
list cab be constructed in different ways, it can be constructed manually, using loops, or lastly list comprehension.

```python
list = [] #1 manually
list = [1, 2, 3]

my_list = [] #2 using loop
for i in range(10):
    my_list.append(i)
print(my_list)

my_list = [number for number in range(10)] #3 list comprehension
print(my_list)
```

```python
square_one_to_ten = []
for i in range(1, 11):
    square_one_to_ten.append(i ** 2)
print(square_one_to_ten)
```

# Recursive function
function that call itself

```python
def f(x):
    if x > 10: # the base case
        return 0
    
    x += 1
    print(x)

    return f(x)

f(0)

'''
f(0) -> x=1 -> print(1) -> f(1)
f(1) -> x=2 -> print(2) -> f(2)
...
f(10) -> x=11 -> print(11) -> f(11)
f(11) -> x=12 -> print(0)
'''
```

```python
def factorial(x):
    for i in range(x+1):
        if x == 1:
            return 1
        return x * factorial(x-1)
factorial(4)
```

```python
def interative_factorial(x):
    result = 1
    for number in range(1, x+1):
        result *= number

    return result
factorial(4)
```

```python
%%timeit
import time
recursive_factorial(50)

t1 = time.time()
result = recursive_factorial(6)
t2 = time.time()
print('Time usage ')
```
