### run time error
**Runtime** is the period when a program is actually running/executing, as opposed to when you’re writing it (coding) or compiling it (turning code into machine code).

Definition: A runtime error occurs while the program is running, and it stops the program from continuing.

This happens when something unexpected occurs during execution, usually due to invalid operations, inputs, or other conditions that the program can’t handle.

#### complie-time error
**Compile-time** is the phase when your code is being converted into a form that the computer can execute before the program actually runs.
This happens before runtime.

Definition: A compile-time error is an error detected by the compiler (or interpreter) before the program runs. These errors stop the program from executing until they are fixed.

```python
def print_phone_num(phone_num):
    assert type(phone_num) == str, 'phone number should be a str'

print_phone_num(6689)
```

```python
address = input('enter your address')
postal_code = address.split()[-1]

def print_postal_code(code):
    assert type(code) == str, f'Postal code need to be string but get {type(code)}'
    assert len(code) == 5, f'postal code needs to have 5 digits'
    print(code)
print_postal_code(postal_code)
```
