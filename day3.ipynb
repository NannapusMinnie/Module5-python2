## Class Method

```python
class Cat:
    name = 'Oscar'
    gender = 'Male'
    age = 5
    weight = 10.4
    color = 'orange'

    def greeting(self):
        return 'Meow!'
    
cat = Cat()
cat.greeting()
```

## _init_constructor

```python
class Cat:
    name = 'Oscar'
    gender = 'Male'
    age = 5
    weight = 10.4
    color = 'orange'

# self is a must included parameter
def _init_(self):
        print('creating an instance here')
        print(1 + 2)
    
cat1 = Cat()
cat2 = Cat()
```

```python
class Cat:
    name = 'Oscar'
    gender = 'Male'

def __init__(self, age, weight, color):
    self.age = age
    self.weight = weight
    self.color = color

cat1 = Cat(age=1, weight=3, color='orange')
cat2 = Cat(age=2, weight=4, color='white')
print(cat1.name, cat1.age, cat1.weight)
print(cat2.name, cat2.age, cat2.weight)
```
