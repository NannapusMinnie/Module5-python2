## Class Method

```python
class Cat:
    name = 'Oscar'
    gender = 'Male'
    age = 5
    weight = 10.4
    color = 'orange'

    def greeting(self):
        return 'Meow!'
    
cat = Cat()
cat.greeting()
```

## __init__constructor

```python
class Cat:
    name = 'Oscar'
    gender = 'Male'
    age = 5
    weight = 10.4
    color = 'orange'

# self is a must included parameter
    def __init__(self):
        print('creating an instance here')
        print(1 + 2)
    
cat1 = Cat()
cat2 = Cat()
```

```python
class Cat:
    name = 'Oscar' # class attribute (same for all object)
    gender = 'Male'

    def __init__(self, age, weight, color):
        self.age = age # instance attribute (different across each object)
        self.weight = weight
        self.color = color

cat1 = Cat(age=1, weight=3, color='orange')
cat2 = Cat(age=2, weight=4, color='white')
print(cat1.name, cat1.age, cat1.weight)
print(cat2.name, cat2.age, cat2.weight)
```

## Magic Methods and special class attributes

```python
# __str__
class Cat:
    def __str__(self) -> str:
        return 'This is a cat'
c = Cat()
print(c)
print(f'Hello, this is a message from Cat class: {c}')

# __repr__
class Cat:
    def __repr__(self) -> str:
        return 'This is another cat'
c = Cat()
print(c)
print(f'Hello, this is a message from Cat class: {c}')
```
